```python
import pandas as pd
```

```python
df = pd.read_csv("E:\\CC GENERAL.csv")
df
```

```python
df.isnull().sum()
```

```python
df["CREDIT_LIMIT"].fillna(df["CREDIT_LIMIT"].mean, inplace=True)
```

```python
df["MINIMUM_PAYMENTS"].fillna(df["MINIMUM_PAYMENTS"].mean, inplace=True)
```

```python
df.isnull().sum()
```

```python
df = pd.get_dummies(df, columns=['CUST_ID'])
```

```python
for col in df.columns:
  print(col, df[col].dtype)
```

```python
float_cols = df.select_dtypes(include=['float']).columns
df = df[float_cols].apply(lambda x: pd.to_numeric(x, downcast='integer'))
```

```python
from sklearn.cluster import AgglomerativeClustering #Importing our clustering algorithm : Agglomerative
model=AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='complete')
clust_labels=model.fit_predict(df)  #Applying agglomerative algorithm with 5 clusters, using euclidean distance as a metric
```

```python
agglomerative=pd.DataFrame(clust_labels)
agglomerative
```

```python
df.corr()
```

```python
import matplotlib.pyplot as plt
fig =plt.figure()
ax = fig.add_subplot(111)
scatter = ax.scatter (df ['BALANCE'] , df ["PURCHASES"] , c= agglomerative[0], s=50)
ax.set_title("Agglomerative Clustering")
ax.set_xlabel("BALANCE")
ax.set_ylabel("PURCHASES")
plt.colorbar(scatter)
```

```python
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(10,7))
plt.title("Customer Dendrograms")
dend=shc.dendrogram(shc.linkage(df, method="complete"))
```

```python
from sklearn.cluster import KMeans  #Importing our clustering algorithm: KMeans
kmeans=KMeans(n_clusters=5, random_state=0)  #Cluster our data by choosing 5 as number of clusters
kmeans.fit(df)
```

```python
labels=pd.DataFrame(kmeans.labels_)
labels
```

```python
kmeans.predict(df)
print(kmeans.cluster_centers_)   #Printing the coordinates of cluster centers.
```

```python
plt.scatter(df["BALANCE"][kmeans.labels_ == 0],          
            df["PAYMENTS"][kmeans.labels_  == 0],s=80,c='magenta',label='Careful')
plt.scatter(df["BALANCE"][kmeans.labels_ == 1],
           df["PAYMENTS"][kmeans.labels_  == 1],s=80,c='yellow',label='Standard')
plt.scatter(df["BALANCE"][kmeans.labels_  == 2],
           df["PAYMENTS"][kmeans.labels_  == 2],s=80,c='green',label='Target')
plt.scatter(df["BALANCE"][kmeans.labels_  == 3], 
           df["PAYMENTS"][kmeans.labels_  == 3],s=80,c='cyan',label='Careless')
plt.scatter(df["BALANCE"][kmeans.labels_  == 4], 
           df["PAYMENTS"][kmeans.labels_  == 4],s=80,c='burlywood',label='Sensible')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 0], s=300, c='red', label = 'Centroids')
plt.title('Clusters of Customers')
plt.xlabel('BALANCE')
plt.ylabel('PAYMENTS')
plt.legend()
plt.show()
```

```python
sum_of_squared_distance = []
k = range(1,15)
for k in k:
    km = KMeans(n_clusters=k)
    km = km.fit(df)
    sum_of_squared_distance.append(km.inertia_)
```

```python
print(sum_of_squared_distance)
```
